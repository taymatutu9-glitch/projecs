```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# ШЄШ№Ш±ЩЉЩЃ Ш§Щ„Щ†Щ…Щ€Ш°Ш¬
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
# ШЄШ¬Щ…ЩЉШ№ Ш§Щ„Щ†Щ…Щ€Ш°Ш¬
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# ШЄШ­Щ…ЩЉЩ„ Ш§Щ„ШµЩ€Ш± Щ…Щ† Ш§Щ„Щ…ШіШ§Ш± Ш§Щ„Щ…Ш­ШЇШЇ
image_path = 'C:\\Users\\DELL\\Desktop\\image'
```

```python
# Ш§Щ„ШЄШ­Щ‚Щ‚ Щ…Щ† Щ€Ш¬Щ€ШЇ Ш§Щ„Щ…Ш¬Щ„ШЇ
if not os.path.exists(image_path):
    print(f"The directory {image_path} does not exist.")
```

```python
images = []
labels = []
```

```python
for filename in os.listdir(image_path):
        if filename.endswith(('.png', '.jpg', '.jpeg')):  # Ш§Щ„ШЄШЈЩѓШЇ Щ…Щ† ШЈЩ† Ш§Щ„Щ…Щ„ЩЃ Щ‡Щ€ ШµЩ€Ш±Ш©
            img = load_img(os.path.join(image_path, filename), target_size=(150, 150))
            img_array = img_to_array(img)
            images.append(img_array)
            label = filename.split('.')[0]  # Ш§ШіШЄШ®Ш±Ш§Ш¬ Ш§Щ„ШЄШіЩ…ЩЉШ© Щ…Щ† Ш§ШіЩ… Ш§Щ„Щ…Щ„ЩЃ
            labels.append(label)
```

```python
if len(images) != len(labels):
        print(f"Mismatch between number of images ({len(images)}) and labels ({len(labels)})")
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
Convert textual labels to integer numbers using LabelEncode 
```

```python
# ШЄШ­Щ€ЩЉЩ„ Ш§Щ„ШЄШіЩ…ЩЉШ§ШЄ Ш§Щ„Щ†ШµЩЉШ© ШҐЩ„Щ‰ ШЈШ±Щ‚Ш§Щ… ШµШ­ЩЉШ­Ш© ШЁШ§ШіШЄШ®ШЇШ§Щ… LabelEncoder
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)
```

```python
# ШЄШЇШ±ЩЉШЁ Ш§Щ„Щ†Щ…Щ€Ш°Ш¬
model.fit(images, labels_encoded, epochs=25)
```

```python
# ШЄЩ‚ЩЉЩЉЩ… Ш§Щ„Щ†Щ…Щ€Ш°Ш¬
loss, accuracy = model.evaluate(images, labels_encoded)
print('Loss:', loss)
print('Accuracy:', accuracy)     
```
